Matplotlib es una biblioteca de Python ampliamente utilizada para la visualización de datos en gráficos y gráficos 2D. La biblioteca Matplotlib proporciona una gran variedad de funciones para crear diferentes tipos de gráficos, como gráficos de líneas, barras, dispersión, histogramas, gráficos de contorno, entre otros. También es posible personalizar muchos aspectos de los gráficos, como colores, estilos de línea, etiquetas, títulos, entre otros.

Más información en la [documentación oficial de Matplotlib](https://matplotlib.org/).




De acuerdo a la página oficial de matplotlib tenemos lo siguiente acerca de la librería:

*Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. Matplotlib makes easy things easy and hard things possible.*
- *Create publication quality plots.*
- *Make interactive figures that can zoom, pan, update.*
- *Customize visual style and layout.*
- *Export to many file formats.*
- *Embed in JupyterLab and Graphical User Interfaces.*
- *Use a rich array of third-party packages built on Matplotlib.*

Al igual que con `numpy` comenzamos importando la librería.


````python
# Esto solo dice importa de la librería matplotlib el módulo pyplot (separado por el punto) con el alias plt.
# Esto significa que usaremos plt para acceder al código de pyplot
import matplotlib.pyplot as plt
import numpy as np
````

A continuación, veremos un ejemplo sencillo de graficar puntos en la recta identidad entre el 0 y el 1 separados por 0.1. 

Primero obtenemos los puntos necesarios para la gráfica usando la siguiente función de `numpy`. 

````python
x=np.arange(start=0,stop=1,step=0.1)
x
````

Ahora graficamos usando la función `scatter` de la librería `plt` que toma como parámetros las coordenadas x y y que deseamos graficar. 

Después de esto usamos la instrucción `plt.show()` para mostrar la  gráfica. Este paso no es necesario si es la última expresión de la celda porque en cualquier celda siempre se regresa la última expresión.

````python
plt.scatter(x,x)
plt.show()
````

La gráfica anterior, aunque útil, no es muy atractiva visualmente. Para tener un estilo no tenemos que configurar cada parte de la gráfica. Existen estilos en la librería que podemos importar como vemos a continuación.

````python
import numpy as np

x = np.array([1,2,3,4,5,6,7,8])

plt.style.use('ggplot') 
plt.scatter(x,x, c = x, cmap = 'plasma')
````

Para más información, pueden ver una lista de los estilos disponibles en esta [página](https://python-charts.com/matplotlib/styles/).

Una vez llamada la función `plt.style.use`  el estilo permanecerá en el resto del cuaderno, a menos que se invoque la función con otro estilo.

*Recordamos que `plt.style.use` quiere decir que dentro de la librería con alias `plt` y el móudlo `style` de dicha librería se encuentra la función `use`*.

De las personalizaciones básicas y requeridas para todas las gráficas se encuentra el etiquetar los ejes y ponerle un título a la gráfica, esto lo podemos obtener de la siguiente forma.

````python
plt.scatter(x,x)
plt.xlabel('Eje x') 
plt.ylabel('Eje y')
plt.title('Titulo de la gráfica')
plt.show()
````

Para graficar líneas tenemos que dar la instrucción `plt.plot`.

````python
## Lineas

# Generamos datos
x = np.linspace(0, 10, 100)
y = x + 1/3*np.sin(x * 2) - np.cos(x) - 1/7 * np.sin(np.sqrt(x)) + np.random.normal(0, 2, 100)

plt.figure(figsize=(10, 5))
plt.plot(x, y)
plt.title('Gráfica con datos simulados')
plt.xlabel('Tiempo')
plt.ylabel('Precio')
````

Otra opción son las gráficas de barras como vemos en el siguiente código:

````python
fig, ax = plt.subplots()

producto = ['producto_1', 'producto_2', 'producto_3', 'producto_4']
counts = [40, 100, 30, 55]
bar_labels = ['red', 'blue', 'green', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:green', 'tab:orange']

ax.bar(producto, counts, label=bar_labels, color=bar_colors)

ax.set_ylabel('Conteo')
ax.set_title('Cantidad de productos')
ax.legend(title='Color del producto')

plt.show()
````
