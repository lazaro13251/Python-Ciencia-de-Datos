```python
import pandas as pd

df = pd.read_csv('./ventas.csv')

print(df.head())
```

```python
alimentos = df[df['Departamento'] == 'Alimentos']
promedioAlimentos = alimentos['Monto_Venta'].mean()
print(f"El promedio de ventas es: {promedioAlimentos}")
```

```python
import pandas as pd
df = pd.read_csv('./ventas.csv')

hogar = df[(df['Departamento'] == 'Hogar') & (df['Fecha'] == '2022-02-21')]

venta = hogar['Monto_Venta'].values[0] if not hogar.empty else None

print(f"Ventas del departamento de hogar es: {venta}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

ventas = pd.read_csv('./ventas.csv')
minVentas = ventas.groupby('Departamento')['Monto_Venta'].min()

plt.figure(figsize=(10, 6))
minVentas.plot(kind='bar')
plt.title('Venta m√≠nima por departamento')
plt.xlabel('Departamento')
plt.ylabel('Monto Venta M√≠nima')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
